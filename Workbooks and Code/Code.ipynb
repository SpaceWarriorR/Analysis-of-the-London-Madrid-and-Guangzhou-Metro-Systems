# London Underground

```python
import networkx as nx
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150

# Create network object from csv file
L = nx.Graph ()
edges = pd.read_csv("London//Edges.csv")
L.add_edges_from(i for i in zip(edges["SourceID"].tolist (), edges["TargetID"].tolist ()))
# nx.write_gexf(G, "London.gexf")
```

## Centralities

### Degree centrality

```python
# Extracting the data
degdict = L.degree()
counts = dict()
for i in degdict:
    counts[i[1]] = counts.get(i[1], 0) + 1
counts = dict(sorted(counts.items()))

# Plotting the result
plt.bar(range(max(list(counts.keys()))), list(counts.values()), tick_label=list(counts.keys()))
plt.title(f'Distribution of degrees in London', fontsize=14)
plt.xlabel('Degree')
plt.ylabel('Frequency')
# plt.show()
```

### Closeness Centrality

```python
# Extracting the data
closenesses = nx.closeness_centrality(L)
values = list(closenesses.values())

# Plotting the result
plt.figure()
sns.kdeplot(values, fill=True, color='blue')
plt.xlim(0, 0.15)
plt.title(f'Density plot of Closeness Centrality in London', fontsize=14)
plt.xlabel('Closeness centrality')
plt.ylabel('Density')
# plt.show()
```

### Betweenness Centrality


```python
# Extracting the data
betweennesses = nx.betweenness_centrality(L)
values = list(betweennesses.values())

# Plotting the result
plt.figure()
sns.kdeplot(values, fill=True, color='blue')
plt.xlim(0, 0.40)
plt.title(f'Density plot of Betweenness Centrality in London', fontsize=14)
plt.xlabel('Betweenness centrality')
plt.ylabel('Density')
# plt.show()
```

### Closeness centrality
